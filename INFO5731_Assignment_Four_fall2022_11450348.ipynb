```python
# Write your code here
import pandas as pd
data=pd.read_csv('https://raw.githubusercontent.com/DakshayiniRayini/dakshayini_INFO5731_Fall2022/main/data1.csv')
print(data.shape)
```

```python
import nltk  
data['Cleaned Data'] = data['Cleaned Data'].str.replace('[^\w\s]','')  
data['Cleaned Data'] = data['Cleaned Data'].apply(lambda a: " ".join(a.lower() for a in a.split()))
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
data['Cleaned Data'] = data['Cleaned Data'].apply(lambda a: " ".join(a for a in a.split() if a not in stop_words))
```

```python
words = []  
from nltk.tokenize import RegexpTokenizer
from gensim import corpora,models
tokenizer = RegexpTokenizer(r'\w+')
for a in pd.Series(data['Cleaned Data']):
  b = tokenizer.tokenize(a)
  words.append(b)
dic = corpora.Dictionary(words)
corp = [dic.doc2bow(word) for word in words]
```

```python
import gensim 
model = gensim.models.ldamodel.LdaModel(corp, num_topics = 10, id2word = dic,passes = 20)
```

```python
print(model.print_topics(num_topics = 10, num_words = 5))
```

```python
from sklearn import metrics

def evaluation(val2_pred, val2_test):
    accuracy = metrics.accuracy_score(val2_pred, val2_test)
    recall = metrics.recall_score(y_pred = val2_pred, y_true = val2_test, pos_label='positive', average='micro')
    precision = metrics.precision_score(y_pred = val2_pred, y_true = val2_test, pos_label='positive', average='micro')
    f1 = 2 * (precision * recall) / (precision + recall)
    print("accuracy= ", accuracy.round(4))
    print("recall=", recall.round(4))
    print("precision=", precision.round(4))
    print("f1 score=", f1.round(4))
```
